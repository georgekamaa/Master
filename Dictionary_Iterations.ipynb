```python
# Getting to see if we can cook based on the Provided ingridients

my_ingridients = ['Eggs', 'Milk', 'Flour']

#The dictionary Iteration

recipes = [
    {'name': 'Pancakes', 'ingredients': ['Eggs', 'Milk', 'Flour']},
    {'name': 'Omelette', 'ingredients': ['Eggs', 'Milk', 'Cheese']},
    {'name': 'Cake', 'ingredients': ['Eggs', 'Milk', 'Flour', 'Sugar']}
]



for recipe in recipes:
    can_cook = True

    for item in my_ingridients:
      if item not in recipe['ingredients']:
        can_cook = False

    if can_cook:
      print(recipe['name'])
```

```python
#Getting to see my type of Movie

my_genres = ['Action', 'Comedy', 'Sci-Fi']

movies = [
    {'title': 'Star Adventure', 'genres': ['Action', 'Sci-Fi']},
    {'title': 'Funny Life', 'genres': ['Comedy', 'Drama']},
    {'title': 'Epic Saga', 'genres': ['Action', 'Comedy', 'Sci-Fi']},
    {'title': 'Romantic Evening', 'genres': ['Romance', 'Drama']}
]


for movie in movies:
  can_watch = True

  for genre in my_genres:
    if genre not in movie['genres']:
      can_watch = False

  if can_watch:
    print(movie['title'])
```

```python
#Getting to know the sport we can play through evaluation of the equipment we have

my_equipment = ['Basketball', 'Tennis Racket', 'Soccer Ball']

sports = [
    {'name': 'Basketball Game', 'equipment': ['Basketball']},
    {'name': 'Soccer Match', 'equipment': ['Soccer Ball']},
    {'name': 'Tennis Match', 'equipment': ['Tennis Racket']},
    {'name': 'Ultimate Sports Challenge', 'equipment': ['Basketball', 'Soccer Ball', 'Tennis Racket']}
]

for sport in sports:
  can_play = True

  for req in my_equipment:
    if req not in sport['equipment']:
      can_play = False
      break

  if can_play:
    print(sport['name'])
```
