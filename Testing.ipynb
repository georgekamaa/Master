```python
class dog: #this is the class
  def __init__(self,name,age): #this is a constructor
    self.name = name
    self.age = age

  def bark(self):#this is the method/fucntion
    print("Woof! My name is", self.name, "and I am", self.age, "old") #this is what the function is actually doing

mydog = dog("Buddy", 3) #this is where we are creating the object/implementing what we had in our class
mydog.bark() #this is where we are calling our object
```

```python
class person: #this is the class
  def __init__(self,name,city): #this is the constructor
    self.name=name #this is 1 varible
    self.city=city #this is another variable

  def greet(self): #this is the fuction/method
    print("Hello my name is", self.name, "and I live in", self.city) #this is what the method will be doing
person1=person("Alex", "London")#this is the object
person1.greet() #this is calling the object
```
